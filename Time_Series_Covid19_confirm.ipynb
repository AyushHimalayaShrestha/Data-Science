# Pivot Table
The pivot table takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['sex'].unique()
```

```python
df.groupby('sex',observed=False)[['total_bill']].sum().reset_index()
```

```python
df.pivot_table(index='sex', columns='smoker',values=['total_bill','tip'], observed=False, aggfunc='sum',margins=True, margins_name='Total')
```

```python
# all cols together

df1=df.pivot_table(index=['sex','smoker'],columns=['day','time'],values=['total_bill'], aggfunc='sum',observed=True)
df1
```

```python
df1.ndim
```

```python
df1=df.pivot_table(index=['sex','smoker'],columns=['day','time'],values=['total_bill'], aggfunc=['sum','mean'],observed=True)
df1
```
