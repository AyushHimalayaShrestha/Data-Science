# 2D line Plot
- Bivariate Analysis
- Categorical-numerical or numerical-numerical
- Usecase - time series data, trending data

```python
# plotting a simple function
price = [48000,50000,54000,57000,47000,45000]
year = [2015,2016,2017,2018,2019.2020,2021]

plt.figure(figsize=(15,5))
plt.plot(year,price)
plt.grid()
plt.title('Price per Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.figure(figsize=[15,5])
plt.plot(batsman['index'],batsman['V Kohli'], label='Virat', color='green', linestyle='dashed', linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'], label='Sharma', color='red', linestyle='dotted', marker='o', markersize=3)
plt.legend(loc='lower center')
plt.grid()
plt.title('Rohit Sharma vs Kohli Career Comparision')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.show()
```

```python
sns.lineplot(batsman, x='index', y='V Kohli', label='V Kohli')
sns.lineplot(batsman, x='index',y='RG Sharma', label ='RG Sharma')
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Score over time')
plt.grid()
plt.legend()
plt.show()
```

# boxPlot
- Univariate analysis
- Numerical
- Use case : statistical distribution, finding outlier

```python
sns.boxplot(x= tips['total_bill'])
```

```python
sns.boxplot(x=tips['tip'])
```

```python
q2=tips['tip'].quantile(.5)
q2
```

```python
tips['tip'].median()
```

```python
q1=tips['tip'].quantile(.25)
q1
```

```python
q3=tips['tip'].quantile(.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
minimum=q1-1.5*iqr
minimum
```

```python
maximum=q3+1.5*iqr
maximum
```

```python
tips['tip']>maximum
```

```python
tips[tips['tip']>maximum]
```

```python
tips[tips['tip']>maximum].count()
```

```python
ind=tips[tips['tip']>maximum].index
ind
```

```python
t=tips.drop(index=ind)
t.describe()
```

```python
sns.boxplot(x=t['tip'])
```

# PieChart
- Unicariate/Bivariate analysis
- Categorical vs Numerical
- Use case :  To find contribution on standard scale.

```python
d=tips.groupby('day', observed=True)[['total_bill']].mean().reset_index()
d
```

```python
plt.pie(d['total_bill'], labels=d['day'])
plt.show()
```

```python
plt.pie(d['total_bill'], labels=d['day'], autopct='%0.2f%%')
plt.legend(bbox_to_anchor=(1,0.5,0.2,0.5))
plt.title('Total bill among Week Day')
plt.show()
```
